```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
root = r"C:\Users\jacob\OneDrive - University of Louisville\Projects\Dissertation\PhaseI\realPatients\journal_Testing"

relatives = [
    #"MESH1_33k\\tracer_conc.out",
    "MESH2_66k\\tracer_conc.out",
    "MESH3_129k\\tracer_conc.out",
    "MESH4_273k\\tracer_conc.out",
    "MESH5_497k\\tracer_conc.out",
    "MESH6_1M\\tracer_conc.out"
]
```

```python
fig, ax = plt.subplots(1, 1)

for mesh in relatives:
    mesh_name = os.path.dirname(mesh)
    file = os.path.join(root, mesh)
    df=pd.read_csv(file, delim_whitespace=True, index_col=0, header=None)
    ax.plot(df[2], df[1], label=mesh_name)
plt.legend()
plt.xlabel('Time [s]')
plt.ylabel('Mass Fraction of Tracer')
plt.xlim(0.02, 0.06)
plt.show()
```

```python
# Courant Analysis
folder = r"C:\\Users\\jacob\\OneDrive - University of Louisville\\Projects\\Dissertation\\PhaseI\\realPatients\\tracer_curves"

fig, ax = plt.subplots(1, 1)

for file in os.listdir(folder):
    df = pd.read_csv(os.path.join(folder, file), delim_whitespace=True, header=None, index_col=0)
    label = file.split('.')
    ax.plot(df.iloc[:, 1], df.iloc[:, 0], label=label[0])

ax.set_xlabel('Time [s]')
ax.set_ylabel('Tracer Mass Fraction')
ax.set_title('Courant Number Analysis')
# ax.set_xlim(0.02, .06)
# ax.set_ylim(0, 0.2e-5)
plt.legend()
plt.show()
```
